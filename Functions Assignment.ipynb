```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list.
def sum_even_numbers(numbers):
    total = 0                # Initialize total to 0
    for num in numbers:      # Loop through each number in the list
        if num % 2 == 0:     # Check if the number is even
            total += num     # Add even number to total
    return total             # Return the sum of even numbers

nums = [2, 5, 8, 11, 14, 17]
result = sum_even_numbers(nums)
print("Sum of even numbers:", result)
```

```python
#2. Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(text):
    return text[::-1]      # Using slicing to reverse the string

# Example usage
word = "Ritika"
result = reverse_string(word)
print("Reversed string:", result)
```

```python
# 3.Implement a Python function that takes a list of integers and returns a new list containing the squares of
# each number.

def square_numbers(numbers):
    squares = []                    # Create an empty list
    for num in numbers:             # Loop through each number
        squares.append(num ** 2)    # Add its square to the list
    return squares                  # Return the new list

# Example usage
nums = [2, 4, 6, 8]
result = square_numbers(nums)
print("List of squares:", result)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200. 
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

# Print prime numbers from 1 to 200
print("Prime numbers from 1 to 200 are:")
for n in range(1, 201):
    if is_prime(n):
        print(n, end=" ")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        num = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return num

# Example use
for i in Fibonacci(10):
    print(i, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(limit):
    for i in range(limit + 1):
        yield 2 ** i

# Example usage
for num in power_of_two(5):
    print(num, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string  def read_file_line_by_line(filename):
with open("sample.txt", "w") as f:
    f.write("First line\n")
    f.write("Second line\n")
    f.write("Third line")

# Step 2: Create generator function
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.rstrip('\n')

# Step 3: Use the generator to read the file
for line in read_file_line_by_line("sample.txt"):
    print(line)
```

```python
#8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple. 
data = [(1, 5), (3, 2), (2, 8), (4, 1)]

# Sorting using lambda function (based on 2nd element)
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
# 9.Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit. 
celsius = [0, 10, 20, 30, 40]

# Convert Celsius to Fahrenheit using map() and lambda
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Temperatures in Fahrenheit:", fahrenheit)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string. 
text = "Hello World"

# Use filter() to remove vowels
result = ''.join(filter(lambda ch: ch.lower() not in 'aeiou', text))

print("String without vowels:", result)
```
